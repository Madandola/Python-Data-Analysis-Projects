## loc and iloc return

*   df.loc[rows, columns name(s)]
*   df.iloc[rows, column index(es)]








```python
df["name"][3]
```

```python
df.loc[0:34 ,"name"]
# return all rows but only name column
```

```python
df.info()
```

```python
df.loc[0:4 ,"name"]
# return first 5 rows but only name column
```

```python
df['calories'].describe()
```

```python
df.loc[0:4, ["name","class"]]
```

```python
df.loc[:, "name"]
# return all rows
# dont write start and end of the range :
```

```python
df.loc[:10, "name"]
# return up to index 10
# dont write start and end of the range :
```

```python
df.loc[10:, "name"]
# return from index 10
# dont write start and end of the range :
```

```python
df.loc[10:, ["name", "id"]]
# return from index 10
# dont write start and end of the range :
```

```python
df.iloc[:, 1]
#1 is index for name column
```

```python
df.iloc[:, [1,2]]
#1 is index for name column
```

## difference between loc and iloc

```python
# df.loc[0:4, ["name","class"]] first 5 rows
# in loc, stopping range is included in the output

df.iloc[0:5, [1,2]]
# in iloc, stopping range is excluded in the output
```

```python
df.iloc[0:5, 0:2]
# 5 rows and 2 columns
```

- Average mark across all students

- Average mark across male and average mark across female students

- Max and min mark

- Max mark for male and max mark for female students

- Average mark for class 'Five'

- Average mark for class 'Six' and male students only

```python
# Average mark across all students
df["mark"].mean()


```

```python
# Average mark across male and average mark across female students

df.loc[df["gender"]=="male", "mark"].mean()
```

```python
# Max and min mark

df["mark"].max()
```

```python
df["mark"].min()
```

```python
# Max mark for male and max mark for female students
df.loc[df["gender"]=="male", "mark"].max()
```

```python
df.loc[df["gender"]=="female", "mark"].max()
```

```python
# Average mark for class 'Five'

df[df["class"]=="Five"]
```

```python
# Average mark for class 'Six' and male students only
df.loc[(df["class"]=="Six") & (df["gender"]=="male"), "mark"].mean()
```
